```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")

df = pd.read_csv("energy_consumption.csv")

df.head()
```

```python
df_missing = df.isnull().sum()
df_missing
```

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

num_cols, cat_cols
```

```python
df_before = df.copy()
```

```python
skews = df[num_cols].skew().sort_values()
skews
```

```python
for col in num_cols:
    if abs(skews[col]) > 0.5:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df_before.isnull(), cbar=False, cmap="Reds")
plt.title("ANTES da Imputação")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap="Greens")
plt.title("DEPOIS da Imputação")
plt.show()
```

```python
outliers_count = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask = (df[col] < lower) | (df[col] > upper)
    outliers_count[col] = mask.sum()

outliers_count
```

```python
df_after_outliers = df.copy()

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_after_outliers = df_after_outliers[
        (df_after_outliers[col] >= lower) & (df_after_outliers[col] <= upper)
    ]
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[num_cols])
plt.title("Antes da Remoção de Outliers")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df_after_outliers[num_cols])
plt.title("Depois da Remoção de Outliers")
plt.xticks(rotation=30)
plt.show()
```

```python
duplicatas = df.duplicated().sum()
df = df.drop_duplicates()
duplicatas
```

```python
skew_values = df[num_cols].skew().sort_values()
skew_values
```

```python
cols_skewed = skew_values[abs(skew_values) > 0.5].index

for col in cols_skewed:
    df[col] = np.log1p(df[col])  # log(1+x)
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
df_encoded.shape[1] - df.shape[1]
```

```python
df_encoded["consumo_por_residente"] = (
    df_encoded["monthly_consumption_kwh"] / df_encoded["num_residents"]
)

df_encoded["densidade_residencial"] = (
    df_encoded["num_residents"] / df_encoded["area_m2"]
)
```

```python
from sklearn.preprocessing import StandardScaler
import joblib

scaler = StandardScaler()
scaled_cols = df_encoded.select_dtypes(include=np.number).columns

df_encoded[scaled_cols] = scaler.fit_transform(df_encoded[scaled_cols])

joblib.dump(scaler, "scaler.pkl")
```
