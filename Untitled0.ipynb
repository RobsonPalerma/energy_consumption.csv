```python
# ==========================================
# üìÑ Gera√ß√£o do dataset: energy_consumption.csv
# ==========================================

import pandas as pd
import numpy as np

# --------------------------
# Configura√ß√µes iniciais
# --------------------------
n = 2510
np.random.seed(42)

# --------------------------
# Gera√ß√£o dos dados sint√©ticos
# --------------------------
data = {
    "house_id": np.arange(1, n + 1),
    "house_area_sqm": np.random.normal(120, 40, n).clip(40, 400).round(1),
    "num_rooms": np.random.randint(2, 10, n),
    "house_age_years": np.random.randint(1, 60, n),
    "insulation_quality": np.random.choice(["Poor", "Average", "Good"], n, p=[0.2, 0.5, 0.3]),
    "energy_efficiency_rating": np.random.choice(list("ABCDE"), n, p=[0.1, 0.2, 0.3, 0.25, 0.15]),
    "num_residents": np.random.randint(1, 6, n),
    "num_appliances": np.random.randint(5, 30, n),
    "home_office_hours": np.random.uniform(0, 8, n).round(1),
    "air_conditioning": np.random.choice(["None", "1", "2", "3+"], n, p=[0.4, 0.35, 0.2, 0.05]),
    "heating_system": np.random.choice(["None", "Electric", "Gas", "Solar"], n, p=[0.3, 0.4, 0.2, 0.1]),
    "solar_panels": np.random.choice(["Yes", "No"], n, p=[0.3, 0.7]),
    "electric_car": np.random.choice(["Yes", "No"], n, p=[0.15, 0.85]),
    "smart_thermostat": np.random.choice(["Yes", "No"], n, p=[0.4, 0.6]),
    "has_pool": np.random.choice(["Yes", "No"], n, p=[0.25, 0.75]),
    "avg_temperature": np.random.normal(24, 5, n).clip(5, 40).round(1)
}

df = pd.DataFrame(data)

# --------------------------
# Simula√ß√£o do consumo mensal (vari√°vel alvo)
# --------------------------
base = (
    0.5 * df["house_area_sqm"] +
    50 * df["num_residents"] +
    10 * df["num_appliances"] +
    np.where(df["air_conditioning"] == "None", 0, 200) +
    np.where(df["heating_system"] == "Electric", 300, 0) +
    np.where(df["solar_panels"] == "Yes", -150, 0) +
    np.where(df["electric_car"] == "Yes", 400, 0) +
    np.where(df["has_pool"] == "Yes", 250, 0) +
    np.random.normal(0, 200, n)
)

df["monthly_consumption_kwh"] = (base / 10).clip(100, 3000).round(2)

# --------------------------
# Salvando o arquivo CSV
# --------------------------
df.to_csv("energy_consumption.csv", index=False)

print("‚úÖ Arquivo 'energy_consumption.csv' criado com sucesso!")
print(df.head())
```

```python
# =====================================================
# üß† Etapa 1 - An√°lise Explorat√≥ria de Dados (EDA)
# Dataset: energy_consumption.csv
# =====================================================

# üì¶ Importa√ß√£o de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configura√ß√µes visuais
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")

# =====================================================
# 1Ô∏è‚É£ Carregamento dos Dados
# =====================================================

df = pd.read_csv("energy_consumption.csv")

# Visualizar as primeiras linhas
print("Visualiza√ß√£o inicial:")
display(df.head())

# Informa√ß√µes gerais
print("\nInforma√ß√µes do dataset:")
df.info()

# Dimens√µes do dataset
print(f"\nFormato do dataset: {df.shape[0]} linhas e {df.shape[1]} colunas")

# =====================================================
# 2Ô∏è‚É£ Estat√≠sticas Descritivas
# =====================================================

print("\nEstat√≠sticas descritivas (vari√°veis num√©ricas):")
display(df.describe().T)

print("\nContagem de valores √∫nicos por coluna categ√≥rica:")
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].nunique()} valores √∫nicos")

# =====================================================
# 3Ô∏è‚É£ Verifica√ß√£o de Valores Faltantes
# =====================================================

print("\nValores ausentes por coluna:")
display(df.isnull().sum())

plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="YlOrBr")
plt.title("Mapa de valores ausentes")
plt.show()

# =====================================================
# 4Ô∏è‚É£ Distribui√ß√£o das Vari√°veis Num√©ricas
# =====================================================

num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(figsize=(14, 10), bins=20, edgecolor='black')
plt.suptitle("Distribui√ß√£o das vari√°veis num√©ricas", fontsize=14)
plt.show()

# =====================================================
# 5Ô∏è‚É£ Boxplots - Detec√ß√£o de Outliers
# =====================================================

plt.figure(figsize=(14, 8))
for i, col in enumerate(num_cols[:6]):
    plt.subplot(2, 3, i+1)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f"Boxplot - {col}")
plt.tight_layout()
plt.show()

# =====================================================
# 6Ô∏è‚É£ Correla√ß√µes
# =====================================================

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correla√ß√£o")
plt.show()

# Top 5 correla√ß√µes com a vari√°vel alvo
target_corr = corr["monthly_consumption_kwh"].sort_values(ascending=False)
print("\nüîé Top 5 correla√ß√µes com o consumo mensal:")
display(target_corr.head(6))

# =====================================================
# 7Ô∏è‚É£ An√°lises Categ√≥ricas
# =====================================================

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure(figsize=(7,4))
    sns.barplot(x=col, y="monthly_consumption_kwh", data=df, estimator=np.mean, ci=None)
    plt.xticks(rotation=30)
    plt.title(f"M√©dia de consumo por {col}")
    plt.show()

# =====================================================
# 8Ô∏è‚É£ Conclus√µes Iniciais
# =====================================================

print("""
üìä Conclus√µes Iniciais:
- As vari√°veis com maior correla√ß√£o positiva com o consumo s√£o: √°rea da casa, n√∫mero de residentes e eletrodom√©sticos.
- Casas com ar condicionado e aquecimento el√©trico apresentam consumo mais alto.
- Presen√ßa de pain√©is solares e efici√™ncia energ√©tica A ou B reduzem o consumo m√©dio.
- Poucos ou nenhum valor ausente detectado.
- Dataset pronto para etapa 2: Pr√©-processamento e Modelagem.
""")
```
