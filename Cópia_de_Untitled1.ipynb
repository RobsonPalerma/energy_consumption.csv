```python
"""
TEMPLATE - ETAPA 3: MODELO BASELINE (Energy Consumption)
========================================================
Objetivo: Criar um modelo baseline de Regress√£o Linear
usando o dataset energy_consumption.csv j√° pr√©-processado.
"""

# ============================================================================
# IMPORTS
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

import joblib

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ============================================================================
# PASSO 1 - CARREGAR DATASET LIMPO
# ============================================================================

print("="*70)
print("ETAPA 3 ‚Äî MODELO BASELINE: REGRESS√ÉO LINEAR")
print("="*70)

# TODO: coloque o caminho onde voc√™ salvou o CSV LIMPO da Etapa 2
df = pd.read_csv("energy_consumption (2) (1).csv")

print(f"\n‚úì Dataset carregado: {df.shape[0]} linhas, {df.shape[1]} colunas")

print("\nPrimeiras linhas:")
print(df.head())

# ============================================================================
# PASSO 2 ‚Äî SEPARAR FEATURES (X) E TARGET (y)
# ============================================================================

TARGET = "monthly_consumption_kwh"

COLS_REMOVE = ["house_id"]   # ID nunca entra no modelo

X = df.drop(columns=[TARGET] + COLS_REMOVE, errors="ignore")
y = df[TARGET]

print("\n‚úì Features utilizadas:")
print(X.columns.tolist())

# Categorias precisam ser codificadas
categoricas = X.select_dtypes(include=["object"]).columns.tolist()
numericas   = X.select_dtypes(include=["int", "float"]).columns.tolist()

print("\nColunas num√©ricas:", numericas)
print("Colunas categ√≥ricas:", categoricas)

# ============================================================================
# PASSO 3 ‚Äî DIVIS√ÉO 60/20/20
# ============================================================================

RANDOM_STATE = 42

# Primeiro: separar TESTE (20%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.20, random_state=RANDOM_STATE
)

# Segundo: Treino (60%) vs Valida√ß√£o (20%) = 75/25 dentro de temp
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=RANDOM_STATE
)

print("\n=== DIVIS√ÉO DOS DADOS ===")
total = len(df)
print(f"Treino: {len(X_train)}  ({len(X_train)/total*100:.1f}%)")
print(f"Val:    {len(X_val)}  ({len(X_val)/total*100:.1f}%)")
print(f"Teste:  {len(X_test)}  ({len(X_test)/total*100:.1f}%)")

# ============================================================================
# PASSO 4 ‚Äî PIPELINE (OneHotEncoder + Regress√£o Linear)
# ============================================================================

preprocessador = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categoricas),
        ("num", "passthrough", numericas),
    ]
)

modelo = Pipeline(steps=[
    ("preprocessamento", preprocessador),
    ("regressor", LinearRegression())
])

print("\nTreinando modelo...")
modelo.fit(X_train, y_train)
print("‚úì Modelo treinado!")

# ============================================================================
# PASSO 5 ‚Äî PREVIS√ïES
# ============================================================================

y_train_pred = modelo.predict(X_train)
y_val_pred   = modelo.predict(X_val)

# ============================================================================
# PASSO 6 ‚Äî M√âTRICAS
# ============================================================================

def metricas(y_true, y_pred, nome=""):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)

    print(f"\n=== {nome} ===")
    print(f"MAE :  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R¬≤  : {r2:.4f}")

    return mse, rmse, mae, r2

print("\n=== M√âTRICAS ===")
train_metrics = metricas(y_train, y_train_pred, "TREINO")
val_metrics   = metricas(y_val, y_val_pred, "VALIDA√á√ÉO")

# Overfitting (RMSE)
gap_rmse = abs(train_metrics[1] - val_metrics[1])
print(f"\nGap de RMSE = {gap_rmse:.3f}")

if gap_rmse < 0.10:
    print("üü¢ Sem overfitting")
elif gap_rmse < 0.30:
    print("üü° Overfitting leve")
else:
    print("üî¥ Overfitting alto")

# ============================================================================
# PASSO 7 ‚Äî GR√ÅFICOS
# ============================================================================

plt.figure(figsize=(8,6))
plt.scatter(y_val, y_val_pred, alpha=0.6)
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], "r--")
plt.title("Predi√ß√µes vs Reais")
plt.xlabel("Real")
plt.ylabel("Previsto")
plt.savefig("pred_real.png", dpi=300)
plt.show()

# Res√≠duos
res = y_val - y_val_pred

plt.figure(figsize=(8,6))
plt.hist(res, bins=30, edgecolor="black")
plt.axvline(0, color="red")
plt.title("Distribui√ß√£o dos Res√≠duos")
plt.xlabel("Res√≠duo")
plt.ylabel("Frequ√™ncia")
plt.savefig("residuos.png", dpi=300)
plt.show()

# ============================================================================
# PASSO 8 ‚Äî SALVAR MODELO
# ============================================================================

joblib.dump(modelo, "baseline_energy.pkl")
print("\n‚úì Modelo salvo: baseline_energy.pkl")

# ============================================================================
# PASSO 9 ‚Äî RESUMO FINAL
# ============================================================================

print("\n=== RESUMO FINAL ===")
print(f"R¬≤ Valida√ß√£o: {val_metrics[3]:.3f}")
print(f"RMSE: {val_metrics[1]:.3f}")
print(f"MAE:  {val_metrics[2]:.3f}")
print("Modelo baseline conclu√≠do!")
```
